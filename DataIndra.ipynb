```python
#memisahkan data menjadi variabel x dan y
X=data.drop(['Outcome'], axis=1)
y=data['Outcome']
```

Dalam analisis klasifikasi, pemisahan data menjadi data X dan data y adalah langkah penting dalam persiapan data. Ini dilakukan untuk memisahkan variabel independen (fitur) dari variabel dependen (OUTCOME) dalam dataset.

Data X (Fitur/Features): Merupakan kumpulan data yang terdiri dari variabel-variabel independen atau fitur-fitur. Fitur-fitur ini adalah atribut-atribut atau informasi-informasi yang digunakan oleh model untuk melakukan prediksi atau klasifikasi.

Data y (Outcome/Labels): Merupakan variabel dependen yang ingin diprediksi oleh model. Ini berisi label atau kelas yang ingin diidentifikasi atau diprediksi berdasarkan fitur-fitur yang ada dalam data X.

Pemisahan ini penting karena:

Melatih Model: Data X digunakan untuk melatih model. Model belajar dari pola-pola yang ada di dalam data X untuk dapat memprediksi atau mengklasifikasikan data dengan tepat.

Validasi dan Evaluasi: Data y (outcome) dibutuhkan untuk menguji seberapa baik model yang sudah dilatih mampu melakukan prediksi. Dengan memisahkan target dari fitur-fitur, kita dapat mengevaluasi seberapa baik model dapat memprediksi label yang benar untuk data baru yang belum pernah dilihat sebelumnya.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler

X_train = MinMaxScaler().fit_transform(X_train)
X_test = MinMaxScaler().fit_transform(X_test)

print(X_train)
```

NAIVE BAYES



```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
```

```python
# fit train set pake Gaussian Naive Bayes
nb.fit(X_train, y_train)
```

```python
# compute the accuracy of train set
nb.score(X_train, y_train)
```

```python
classif_name = ['NB']
y_predict = {}
y_predict['NB'] = nb.predict(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score # Import accuracy_score
nb = GaussianNB()

# fit train set pake Gaussian Naive Bayes
nb.fit(X_train, y_train)

# compute the accuracy of train set
nb.score(X_train, y_train)

classif_name = ['NB']
y_predict = {}
y_predict['NB'] = nb.predict(X_test)


for name in classif_name:
    print('{0} accuracy = {1:.4f}'.format(name, accuracy_score(y_test, y_predict[name]))) # Now accuracy_score is defined
```

Proses membuat model dengan Neural Network

```python
#Menguji parameter yang paling optimum untuk model

from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV

from sklearn.model_selection import StratifiedKFold

cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=100)

# Tentukan parameter yang ingin diuji
parameters = {
    'hidden_layer_sizes': [(50,), (100,), (50, 30), (100, 50)],
    'learning_rate_init': [0.001, 0.01, 0.1],
    'max_iter': [100, 200, 300]
}

# Buat model MLPClassifier
clf = MLPClassifier(random_state=100)

# Inisialisasi GridSearchCV
grid_search = GridSearchCV(clf, parameters, cv=cv_strategy)
# cv merupakan jumlah lipatan validasi silang (cross-validation)

# Lakukan pencarian grid pada data training
grid_search.fit(X, y)
```

```python
# Tampilkan parameter terbaik
print("Parameter terbaik:", grid_search.best_params_)
```

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(hidden_layer_sizes=(50,),learning_rate_init=0.1,max_iter=100)

clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

Hasil dari laporan klasifikasi tersebut memberikan ringkasan performa model pada data uji penyakit Parkinson dengan menggunakan berbagai metrik evaluasi seperti precision, recall, dan f1-score. Berikut adalah interpretasi dari setiap metrik yang tercantum dalam laporan:

Precision:


- 0 (Kelas Negatif/Non-Parkinson): Dari semua prediksi yang dilakukan sebagai kelas negatif, sekitar 83% di antaranya adalah benar-benar kelas negatif.
- 1 (Kelas Positif/Parkinson): Sekitar 77% dari semua prediksi yang dilakukan sebagai kelas positif adalah benar.

Recall:


- 0 (Kelas Negatif/Non-Parkinson): Dari semua sampel yang sebenarnya adalah kelas negatif, model mampu mengenali sekitar 13% dari mereka.
- 1 (Kelas Positif/Parkinson): Dari semua sampel yang sebenarnya adalah kelas positif, model dapat mengenali sekitar 99% dari mereka.

F1-Score:


- F1-score untuk kelas 0 adalah 0.23, yang cukup rendah karena recall yang sangat rendah.
- F1-score untuk kelas 1 adalah 0.87, yang cukup tinggi berkat recall yang sangat baik.

Accuracy:

- Akurasi model adalah sekitar 78%, yang mengindikasikan persentase prediksi benar secara keseluruhan.

Interpretasi:
- Meskipun akurasi secara keseluruhan adalah 78%, melihat pada recall untuk kelas 0 (0.13) menunjukkan bahwa model memiliki performa yang sangat rendah dalam mengidentifikasi kasus negatif (non-Parkinson). Model ini memiliki bias yang sangat kuat untuk memprediksi kelas positif (Parkinson), yang tercermin dari recall 0.99 untuk kelas 1.

Dalam konteks medis, model ini akan mengklasifikasikan hampir semua pasien sebagai penderita Parkinson (hampir semua kasus positif terdeteksi), tetapi banyak pasien sehat juga akan salah diklasifikasikan sebagai penderita Parkinson (banyak false positive). Model ini akan berguna jika tujuan utamanya adalah memastikan tidak ada kasus Parkinson yang terlewatkan, tetapi memerlukan tes tambahan untuk mengonfirmasi diagnosis karena tingginya false positive.
